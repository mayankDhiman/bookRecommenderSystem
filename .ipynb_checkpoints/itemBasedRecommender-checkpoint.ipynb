```python
def btao(naam): 
    userRating = []
    idx = books.bookTitle[books.bookTitle == naam].index.values.astype(int)[0]
    isbn = books.ISBN.loc[idx]
    userRating = ratings_matrix[isbn]
    userRating = ratings_matrix.corrwith(
        userRating)
    userRating = pd.DataFrame(userRating) 
    userRating.columns = ['corr']
    userRating.dropna(inplace = True)
    userRating.drop_duplicates(subset='corr', inplace = True)
    userRating.sort_values(by = 'corr', ascending = False, inplace = True)
    return userRating
```
